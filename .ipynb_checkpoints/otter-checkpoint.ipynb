```python
import pandas as pd
import seaborn as sns
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
sns.set_theme()
```

```python
otter = pd.read_csv("pb.csv")
#remove NAs
otter1 = otter.dropna()
```

```python
otter1
```

```python
sns.relplot(
    data = otter1,
    x="Emissions",
   # x="UWC",
    y="BonePb",
)
```

```python
import matplotlib.pyplot as plt

from plothist import make_hist, plot_hist
```

```python
#distribution of BoneBp 

h = make_hist(otter1["BonePb"])
fig, ax = plt.subplots()

plot_hist(h, ax=ax)

ax.set_xlabel("BonePb")
ax.set_ylabel("Entries")

median_BonePb = otter1['BonePb'].median()
ax.axvline(median_BonePb, linestyle="--", linewidth=1, color="red")
```

```python
#label high/low bonepb

median_BonePb = otter1['BonePb'].median() 
otter1['High_Pb'] = np.where(otter1['BonePb'] > median_BonePb, 1, 0)
```

```python
otter1.head
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(otter1['SedPb'], otter1['BonePb'], c=otter1['High_Pb'], cmap='coolwarm', alpha=0.5)
plt.xlabel('SedPb')
plt.ylabel('BonePb')
plt.title('relationships between')
plt.show()
```

```python
# select numeric variables
X = otter1[['Wt','L','Emissions','pH','Rain','Wind','SedCa','SedPb']]
y = otter1['High_Pb']
```

```python
# Standardize features (allow different ranges of numbers to be compared fairly)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Divide training set and test set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
# Train a logistic regression model
model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)
print("Prediction results of the first 5 test samplesï¼š", y_pred[:5])
print("Top 5 real resultsï¼š", y_test.values[:5])
```

```python
# Calculate model accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model accuracyï¼š{accuracy:.2f}ï¼ˆProportion of correct predictionsï¼‰")
```

```python
# Draw confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['low BonePb', 'high BonePb'])
disp.plot(cmap='Blues')
plt.title('matrix')
plt.show()
```

```python
# See which features are important
feature_importance = pd.DataFrame({
    'features': ['(Wt)','(L)','(Emissions)', '(Rain)', '(Ph)','(Wind)', '(SedCa)', '(SedPb)'],
    'weights': model.coef_[0]
})
print("importance of featuresï¼š")
print(feature_importance)
```
